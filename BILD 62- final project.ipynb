# Project Description
## Goal of project
This project aims to use our codes to determine the most ideal acceptor fluorophore given a donor fluorophore. 
Fluorescent proteins have become fundamentally important to science for their proven versatility and significant benefits, such as serving as markers for gene expression, biosensors, optical highlighters, and enabling imaging previously impossible. Since fluorescent proteins have continued to be helpful in new applications of scientific research, the motivation for this project is that it will help us figure out the matching fluorescents more efficiently, given a list of fluorescent candidates. For example, scientists can use our code to find the best fluorescent that matches the original fluorescent more efficiently when they discover a new fluorescent protein. 
We will have a table containing a list of acceptor fluorophores (extracted from Table 1, in the link below) with their corresponding wavelength range required for input and a donor fluorophore pre-chosen to act as a newly discovered protein. Then, we are going to import the table into python through Pandas. And then, we are going to generate a certain number of graphs corresponding to the number of items in the table. Within the first graph, two different normal distributions will represent the first acceptor’s input range and the donor’s output range. These graphs will be automatically generated by creating our first new function using a loop code that loops through all the acceptor fluorophores. Then, we will write a piece of code, the second new function, that determines the ideal acceptors by checking whether the overlapping area meets the minimum requirement pre-set by us and prints out the following:" The nth acceptor is/ is not a perfect choice!" for each of the acceptors in the list. Lastly, our third new function will consist of a code that compares the overlapping areas generated by those ideal choices and will print out:" This is the most ideal!" for the acceptor with the largest overlap.
